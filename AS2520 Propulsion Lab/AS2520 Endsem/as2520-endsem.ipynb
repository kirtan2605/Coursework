```python
initial_diameter = [0.009, 0.017, 0.025, 0.033, 0.041]    # in m
final_diameter = [0.017, 0.025, 0.033, 0.041, 0.049]    # in m

Ap = []    # in m^2 in e-5
Gox = []    # in kg/s.m^2
rdot = []    # in m/s
mdotf = []    # in mm

# regression relation constants
a = 0.116
n = 0.62

# fuel properties
rho_fuel = 900  #  kg/m^3

for i in [0,1,2,3,4]:
    Ap_val = 0.25 * np.pi * ((initial_diameter[i] + final_diameter[i])*0.5)**2
    Ap.append(Ap_val*1e5)
    Gox_val = 0.030/Ap_val
    Gox.append(Gox_val)
    rdot_val_1 = a*(Gox_val**n)  # rdot in mm/s
    rdot_val_2 = rdot_val_1*1e-3   #rdot on m/s
    rdot.append(rdot_val_2*1e3)
    mdotf_val = Ap_val*rho_fuel*rdot_val_2
    mdotf.append(mdotf_val)

print("Ap in e-5")
print(Ap)
print("Gox")
print(Gox)
print("r_dot in m/s in 1e-3")
print(rdot)
print("mdotf")
print(mdotf)
```

```python
Lg = 0.134   # in m
delta_m = []

for j in [0,1,2,3,4]:
    delta_m_val = np.pi * rho_fuel * Lg * (((final_diameter[j])**2) - ((initial_diameter[j])**2 )) * 0.25
    delta_m.append(delta_m_val)
    
print("delta_m")
print(delta_m)
```

```python
delta_t = []

for j in [0,1,2,3,4]:
    delta_t_val = delta_m[j]/mdotf[j]
    delta_t.append(delta_t_val)
    
    
print("delta_t")
print(delta_t)
```

```python
O_by_F = []

m_dot_O = 0.03

for j in [0,1,2,3,4]:
    O_by_F_val = m_dot_O/mdotf[j]
    O_by_F.append(O_by_F_val)
    
       
print("O/F")
print(O_by_F)
```

```python
# print(mdotf)

mdotf_array = np.array(mdotf)

#print(mdotf_array)
m_total = mdotf_array + 0.030

At = np.pi*0.25*((0.008)**2)

c_star = 327.22207

Pb = 101325   # in Pa

Pc = (c_star/At)*m_total + Pb   # in Pa

print("cstar")
print(c_star)
print(Pc/101325)   # in atm

print("Pe")
Pe = (Pc/101325)*0.564450668
print(Pe)
```

```python
CF = []
Pc = Pc/101325
for i in Pc:
    print(i)
    CF_val = 1.24656759 + 1/i
    CF.append(CF_val)
    
print(CF)
```

```python
Thrust = []

for i in [0,1,2,3,4]:
    T_val = c_star * m_total[i] * CF[i]
    Thrust.append(T_val)
    
print(Thrust)
```
