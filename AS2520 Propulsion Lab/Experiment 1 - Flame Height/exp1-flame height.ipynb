```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Case-1: LPG (10mm)
x1 = np.array([6.05, 10, 15, 20, 25, 30, 35])
x1_corrected = x1 * 0.722    # correction for LPG gas
x1_data = x1_corrected.tolist()
y1_data = [70, 75, 110, 120, 130, 140, 150]

polynomial_coeff=np.polyfit(x1_data,y1_data,3)
xnew=np.linspace(0,30,100)
ynew=np.poly1d(polynomial_coeff)
plt.plot(xnew,ynew(xnew),x1_data,y1_data,'o')
plt.title("LPG (10 mm diameter)")
plt.xlabel("Corrected Flow Rate (in LPM)")
plt.ylabel("Flame Height (in cm)")

plt.show()

#Case-1: LPG (10mm) - Correct Demarcation
x1 = np.array([5, 10, 15, 20, 25, 30, 35])
x1_corrected = x1 * 0.722    # correction for LPG gas
x1_data = x1_corrected.tolist()
y1_data = [70, 75, 110, 120, 130, 140, 150]

polynomial_coeff=np.polyfit(x1_data,y1_data,3)
xnew=np.linspace(0,30,100)
ynew=np.poly1d(polynomial_coeff)
plt.plot(xnew,ynew(xnew),x1_data,y1_data,'o')
plt.title("LPG (10 mm diameter)")
plt.xlabel("Corrected Flow Rate (in LPM)")
plt.ylabel("Flame Height (in cm)")

plt.show()
```
